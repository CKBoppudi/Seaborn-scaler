# Calculating Inter quartile ranges.

# Inter Quartile Range (IQR) for setosa = q3 - q1 = 3.8 - 3.3 = 0.5

# Inter Quartile Range (IQR) for versicolor = q3 - q1 = 3 - 2.5 = 0.5

# Inter Quartile Range (IQR) for virginica = q3 - q1 = 3.2 - 2.7 = 0.5

# ############--------- Use cases of boxplot ----------#######
# 1. Stock markets.
# 2. To detect range.
# 3. To detect outliers.
# 4. To identify IQR (Inter quartile range) --IQR is the range between 25th percentile and 75th percentile

# ##########---------- Outlier Formula ----------##########

# Outlier lower bound ---- Q1 - (IQR * 1.5)
# points below the lower bound are lower outliers
# Q1 = 1st Quartile
# IQR = Q3 - Q1

# Outlier upper bound ---- Q3 +( IQR * 1.5)
# points above the upper bound are upper outliers
# Q3 = 3rd Quartile
# IQR = Q3 - Q1
